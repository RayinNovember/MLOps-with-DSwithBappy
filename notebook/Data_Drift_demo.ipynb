```python
import pandas as pd
import json
```

```python
from evidently.legacy.report import Report
from evidently.legacy.metric_preset import DataDriftPreset, TargetDriftPreset
#from evidently.profile_sections import DataDriftProfileSection
```

```python
data_url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
boston_frame = pd.read_csv(data_url)
```

```python
data_drift_dashboard = Report(metrics=[DataDriftPreset()])
data_drift_dashboard.run(reference_data=boston_frame[:200], current_data=boston_frame[200:])
data_drift_dashboard.show()                    
data_drift_dashboard.save_html("report.html")  
```

```python
json_report = json.loads(data_drift_dashboard.json())
```
