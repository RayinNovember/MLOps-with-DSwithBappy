### Check Null Values

```python
features_with_na = [feature for feature in df.columns if df[feature].isnull().sum()>=1]
len(features_with_na)
for feature in features_with_na:
    print(feature,np.round(df[feature].isnull().mean()*100,5),"% missing values")
```
